```python
import pandas as pd

#cargar los datos a un DataFrame
archivo_excel = "C:\\Users\\ssgon\\OneDrive - Universidad de los andes\\2025-1\\Analitica\\proyecto_aptos\\datos_apartamentos_rent.xlsx"
df = pd.read_excel(archivo_excel)
print(df.head())
```

```python
#estadisticas descriptivas

df.describe()
```

Histogramas

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Histogramas de todas las variables numéricas
df["bathrooms"].hist(figsize=(6, 5), bins=30, grid=False, edgecolor="black")
plt.xlabel("Número de baños")
plt.ylabel("Frecuencia")
plt.suptitle("Histogramas de Variables Numéricas", fontsize=14)
plt.show()

df["bedrooms"].hist(figsize=(6, 5), bins=30, grid=False, edgecolor="black")
plt.xlabel("Número de habitaciones")
plt.ylabel("Frecuencia")
plt.show()

df["price"].hist(figsize=(6, 5), bins=30, grid=False, edgecolor="black")
plt.xlim(0,52500)
plt.xlabel("Precio (USD)")
plt.ylabel("Frecuencia")
plt.show()

df["square_feet"].hist(figsize=(6, 5), bins=30, grid=False, edgecolor="black")
plt.xlim(0,40000)
plt.xlabel("Tamaño del apartamento (pies cuadrados)")
plt.ylabel("Frecuencia")
plt.show()
```

Boxplots

```python
#Diagrama de cajas para variables discretas
plt.figure(figsize=(12, 5))
sns.boxplot(x="category", y="price", data=df, showfliers=False)
plt.xlabel("Categoría")
plt.ylabel("Precio (USD)")
plt.title("Distribución del Precio por Categoría")
plt.show()
```

```python
plt.figure(figsize=(12, 5))
sns.boxplot(x="has_photo", y="price", data=df, showfliers=False)
plt.xlabel("Tiene foto")
plt.ylabel("Precio (USD)")
plt.title("Distribución del Precio por presencia de fotos")
plt.show()
```

```python
plt.figure(figsize=(12, 5))
sns.boxplot(x="pets_allowed", y="price", data=df, showfliers=False)
plt.xlabel("Que mascotas se permiten")
plt.ylabel("Precio (USD)")
plt.title("Distribución del Precio según las mascotas permitidas")
plt.show()
```

```python
plt.figure(figsize=(12, 5))
sns.boxplot(x="state", y="price", data=df, showfliers=False)
plt.xlabel("Estado")
plt.ylabel("Precio (USD)")
plt.title("Distribución del Precio según el estado")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12, 5))
sns.boxplot(x="bathrooms", y="price", data=df, showfliers=False)
plt.xlabel("Número de baños")
plt.ylabel("Precio (USD)")
plt.title("Distribución del Precio según el Número de baños")
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(12, 5))
sns.boxplot(x="bedrooms", y="price", data=df, showfliers=False)
plt.xlabel("Número de habitaciones")
plt.ylabel("Precio (USD)")
plt.title("Distribución del Precio según el número de habitaciones")
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(12, 5))
sns.boxplot(x="source", y="price", data=df, showfliers=False)
plt.xlabel("Origen del anuncio")
plt.ylabel("Precio (USD)")
plt.title("Distribución del Precio según el origen del anuncio")
plt.xticks(rotation=45)
plt.show()
```

Diagramas de dispersión

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x="square_feet", y="price", data=df, color="blue", alpha=0.5)
plt.xlabel("Tamaño del apartamento (square_feet)")
plt.ylabel("Precio (USD)")
plt.title("Diagrama de Dispersión: Precio vs. Tamaño")
plt.show()

plt.figure(figsize=(8, 5))
sns.scatterplot(x="longitude", y="latitude", data=df, color="blue", alpha=0.5)
plt.xlabel("Longitud")
plt.ylabel("Latitud")
plt.title("Diagrama de Dispersión: Latitud vs. Longitud")
plt.show()

plt.figure(figsize=(8, 5))
sns.scatterplot(x="latitude", y="price", data=df, color="blue", alpha=0.5)
plt.xlabel("Latitud")
plt.ylabel("Precio (USD)")
plt.title("Diagrama de Dispersión: Precio vs. Latitud")
plt.show()

plt.figure(figsize=(8, 5))
sns.scatterplot(x="longitude", y="price", data=df, color="blue", alpha=0.5)
plt.xlabel("Longitud")
plt.ylabel("Precio (USD)")
plt.title("Diagrama de Dispersión: Precio vs. Longitud")
plt.show()
```

```python
# Correlación entre variables numéricas
var_numericas = ['bathrooms', 'bedrooms', 'square_feet', 'latitude', 'longitude', 'price']
matriz_cor = df[var_numericas].corr()

# Mapa de calor de correlaciones
plt.figure(figsize=(8,6))
sns.heatmap(matriz_cor, annot=True, cmap='coolwarm')
plt.title("Matriz de Correlación entre Variables")
plt.show()

#Dispersión entre variables
var_interes = ['bathrooms', 'bedrooms', 'square_feet', 'price']
sns.pairplot(df[var_interes])

```
