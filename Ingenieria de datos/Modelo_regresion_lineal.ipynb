```python
import pandas as pd
```

```python
datos_brutos = pd.read_csv("datos_apartamentos_rent.csv", encoding="latin-1", sep=";")
datos_brutos.head()
```

```python
datos_limpios=pd.read_csv("datos_limpios.csv", encoding="latin-1", sep=";")
datos_limpios.head()
```

```python
datos_finales=pd.DataFrame()
```

```python
datos_finales['Mascotas_permitidas'] = datos_brutos['pets_allowed']
datos_finales["Mascotas_permitidas"] = datos_finales["Mascotas_permitidas"].fillna("Ninguno")
```

```python
# Crear un diccionario con índices únicos
unique_types = {tipo: idx for idx, tipo in enumerate(datos_finales["Mascotas_permitidas"].unique())}
# Aplicar la codificación
datos_finales["Codificacion_Mascotas_Permitidas"] = datos_finales["Mascotas_permitidas"].map(unique_types)
```

```python
datos_finales.head(259)
```

```python
valores_unicos=datos_brutos["source"].unique()
valores_unicos
```

```python
datos_finales['Source'] = datos_brutos['source']
# Crear un diccionario con índices únicos
unique_types = {tipo: idx for idx, tipo in enumerate(datos_finales["Source"].unique())}
# Aplicar la codificación
datos_finales["Codificacion_Source"] = datos_finales["Source"].map(unique_types)
```

```python
datos_finales.head(259)
```

```python
datos_finales['Tiene_foto'] = datos_brutos['has_photo']
# Crear un diccionario con índices únicos
unique_types = {tipo: idx for idx, tipo in enumerate(datos_finales["Tiene_foto"].unique())}
# Aplicar la codificación
datos_finales["Codificacion_Tiene_foto"] = datos_finales["Tiene_foto"].map(unique_types)
datos_finales.head(20)
```

```python
datos_finales['Categoria'] = datos_brutos['category']
# Crear un diccionario con índices únicos
unique_types = {tipo: idx for idx, tipo in enumerate(datos_finales["Categoria"].unique())}
# Aplicar la codificación
datos_finales["Codificacion_Categoria"] = datos_finales["Categoria"].map(unique_types)
datos_finales.head(10)
```

```python
datos_finales = datos_finales.drop(columns=["Mascotas_permitidas", "Source", "Tiene_foto","Categoria"])
```

```python
datos_finales[["amenities_count", "state", "cityname","square_feet","price","bedrooms"]]=datos_limpios[["amenities_count", "state", "cityname","square_feet","price","bedrooms"]]
```

```python
datos_finales["bathrooms"] = datos_limpios["bathrooms"].str.replace(",", ".").astype(float)
```

```python
datos_finales.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
datos_finales_scaled = pd.DataFrame(scaler.fit_transform(datos_finales), columns=datos_finales.columns)
```

```python
datos_finales_scaled
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

# Agregar una columna de unos (intercepto)
datos_finales_scaled["intercept"] = 1  

# Calcular VIF para cada variable
vif_data = pd.DataFrame()
vif_data["Variable"] = datos_finales_scaled.columns
vif_data["VIF"] = [variance_inflation_factor(datos_finales_scaled.values, i) for i in range(datos_finales_scaled.shape[1])]

print(vif_data)
```
